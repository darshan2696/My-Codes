```python
def duration():                                 #determines the number of songs that can be played in the given break duration
    r_1=0   
    itr=0
    global k,x
    for i in k:
        url = (i)
        video = pafy.new(url)
        r=video.length
        itr +=1
        if r != r_1:                          
            r_1 = r+r_1     
            if r_1>dur:
                x=itr
                shuffle(k)
                k=k[:x]
                break
            

       
```

```python
def video_duration(a):
    url = (k[a])
    video = pafy.new(url)
    return (video.length)                        #returns the duration of the youtube video
```
